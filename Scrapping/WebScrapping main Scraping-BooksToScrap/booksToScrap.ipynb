```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
from tqdm import tqdm
```

```python

link = "https://books.toscrape.com"
    
res = requests.get(link)
soup = BeautifulSoup(res.text , "html.parser")
data = []
    
for i in soup.find_all("li" , class_ = "col-xs-6 col-sm-4 col-md-3 col-lg-3"):
    data.append([
        (i.find_all("a")[-1].get("title")),
        (i.find("p" , class_ = "star-rating" ).get("class")[-1]),
        (i.find("p", class_ = "price_color").text[1:]),
        (i.find("p" , class_ = "instock availability").text.strip()),
        (link+i.find("img", class_ = "thumbnail").get("src")),
        (link+i.find("a").get("href")),])
```

```python
data
```

```python
df = pd.DataFrame(data , columns = ["Book_Name" , "Rating" , "Price" , "Availablity" , "Image_link","page_link"])
```

```python
df.to_csv("first_page_data.csv")
```

## multipage scraping

```python
data_2 = []
j = 0

for i in tqdm(range(1,51)):
    link = "https://books.toscrape.com/catalogue/page-"+str(i)+".html"
    
    res = requests.get(link)
    soup = BeautifulSoup(res.text , "html.parser")
    
    
    for i in soup.find_all("li" , class_ = "col-xs-6 col-sm-4 col-md-3 col-lg-3"):
        data_2.append([
        (i.find_all("a")[-1].get("title")),
        (i.find("p" , class_ = "star-rating" ).get("class")[-1]),
        (i.find("p", class_ = "price_color").text[1:]),
        (i.find("p" , class_ = "instock availability").text.strip()),
        (link[:26]+i.find("img", class_ = "thumbnail").get("src")[2:]),
        (link[:37]+i.find("a").get("href"))])


 
```

```python
data = []
for link in tqdm(df["page_link"]):
    res = requests.get(link)
    soup = BeautifulSoup(res.text , "html.parser")

    genre =  soup.find("ul" , class_ = "breadcrumb").find_all("a")[2].text
    title = soup.find("h1").text
    
    book = soup.find("table" , class_ = "table table-striped").find_all("td")
    upc = book[0].text
    priceEx = book[2].text[1:]
    priceIn = book[3].text[1:]
    tax = book[4].text[1:]
    avl = soup.find("p", class_ = "instock" ).text.strip()
   
    reviews = book[6].text
    data.append([title,genre,upc , priceEx , priceIn , tax , avl , reviews])
```
